Q 1. create a function that calculates the final salary of an employee after adding bonus and deducting tax.

```python
def calculate_final_salary(basic_salary, bonus, tax):
  final_salary = basic_salary + bonus - tax
  return final_salary

salary = calculate_final_salary(50000, 5000, 3000)

print("Final Salary :" , salary)

```

Q 2. Create a function that accepts marks of multiple subjects using *args and calculates total, average, and grade.

```python
def calculate_result (*marks):
  total = sum(marks)
  average = total / len(marks)

  if average >= 90:
    grade = "A+"
  elif average >= 80:
    grade = "A"
  elif average >= 70:
    grade = "B"
  elif average >= 60:
    grade = "C"
  elif average >= 50:
    grade = "D"
  else:
    grade = "F"

  print("Total Marks: ", total)

  print("Average Marks: ", average)

  print("Grade: ", grade)

calculate_result (85, 90, 78, 88, 92)

```
