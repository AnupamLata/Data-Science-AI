#Operators

1. Arithmetic Operator

```python
a = 15
b = 4

print("Addition:", a+b)
print("Subtraction:", a-b)
print("Multiplication:", a*b)
print("Division:", a / b)
print("Floor Division:", a // b)
print("Modules:", a % b)
print("Exponentiation:", a ** b)

```

2. Comparision Operator

```python
a = 13
b = 33

print(a > b)
print(a < b)
print(a == b)
print(a != b)
print(a >= b)
print(a <= b)
```

3. Logical Operators

```python
a = True
b = False
print(a and b)
print(a or b)
print(not a)
```

4. Bitwise operator

```python
a = 10
b = 4
print(~a)
print(a << b)
print(a & b)
print(a ^ b)
print(a | b)
print(a >> b)
```

5. Assignment Operators

```python
a = 10
b = a
print(b)
b += a
print(b)
b -= a
print(b)
b *= a
print(b)
b <<= a
print(b)
```

6. Identity Operator

```python
a = 10
b = 20
c = a

print(a is not b)
print(a is c)

```

7. Membership Opertors

```python
x = 24
y = 20
my_list = [10, 20, 30, 40, 50]

if (x not in my_list):
  print("x is NOT present in given list")
else:
  print("x is present in given list")
if (y in my_list):
  print("y is present in given list ")
else:
  print("y is Not present in given list")
```

8. Ternary Operator

```python
a, b = 10, 20
min = a if a < b else b

print(min)
```
